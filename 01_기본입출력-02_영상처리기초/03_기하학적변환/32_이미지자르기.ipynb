# 이미지 자르기

## 영역을 잘라서 새로운 윈도우(창)에서 표시

```python
import cv2
img_file = '../cv_images/img.jpg'
img = cv2.imread(img_file)

# img.shape : (390, 640, 3)  => 이 이미지보다 작게 자름
crop = img[100:200, 300:400] # ... 코드입력  # 세로기준 100:200, 가로기준 300:400

cv2.imshow('img', img)
cv2.imshow('crop', crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
img_plt= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_plt)
```

## 영역을 잘라서 기본 윈도우에 표시

```python
import cv2
img_file = '../cv_images/img.jpg'
img = cv2.imread(img_file)

dst = img.copy() 
crop = img[100:200, 200:400] # 세로기준 100:200, 가로기준 300:400
dst[100:200, 400:600] = crop       #... 코드입력  # 세로기준 100:200, 가로기준 400:600에 crop을 넣음

cv2.imshow('img', img)
cv2.imshow('dst', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
