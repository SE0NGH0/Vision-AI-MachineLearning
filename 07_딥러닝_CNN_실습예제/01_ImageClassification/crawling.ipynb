```python
from icrawler.builtin import GoogleImageCrawler
import os

BASE_PATH = "C:\\Users\\intel09\\Documents\\07_딥러닝_CNN_실습예제\\07_딥러닝_CNN_실습예제\\data\\"

def download_images(keyword, num_images):
    # 키워드별 폴더 생성
    save_dir = os.path.join(BASE_PATH, keyword)
    os.makedirs(save_dir, exist_ok=True)

    # 크롤러 설정
    crawler = GoogleImageCrawler(storage={'root_dir': save_dir})
    crawler.crawl(keyword=keyword, max_num=num_images)
    print(f"[완료] '{keyword}' 키워드로 {num_images}장의 이미지가 다운로드되었습니다.")

def run_image_downloader():
    try:
        num_sets = int(input("몇 세트를 다운로드하시겠습니까? > "))
    except ValueError:
        print("숫자를 입력해주세요.")
        return

    for i in range(num_sets):
        keyword = input(f"[{i+1}/{num_sets}] 검색할 이미지 키워드를 입력하세요 > ").strip()
        try:
            num_images = int(input(f"[{keyword}] 다운로드할 이미지 개수 > "))
        except ValueError:
            print("숫자를 입력해주세요.")
            continue

        download_images(keyword, num_images)

if __name__ == "__main__":
   run_image_downloader()
```
